```python
import gspread
from oauth2client.service_account import ServiceAccountCredentials

from google.colab import drive
import matplotlib.pyplot as plt

drive.mount('/content/drive')
```

```python
import numpy as np

adadelta_acc = np.genfromtxt('582 HW 4 data - adadelta acc.csv', delimiter=',')
adadelta_loss = np.genfromtxt('582 HW 4 data - adadelta loss.csv', delimiter=',')
adam_acc = np.genfromtxt('582 HW 4 data - adam acc.csv', delimiter=',')
adam_loss = np.genfromtxt('582 HW 4 data - adam loss.csv', delimiter=',')
baseline_acc = np.genfromtxt('582 HW 4 data - baseline acc.csv', delimiter=',')
baseline_loss = np.genfromtxt('582 HW 4 data - baseline loss.csv', delimiter=',')
net2_acc = np.genfromtxt('582 HW 4 data - net2 acc.csv', delimiter=',')
net2_loss = np.genfromtxt('582 HW 4 data - net2 loss.csv', delimiter=',')
net3_acc = np.genfromtxt('582 HW 4 data - net3 acc.csv', delimiter=',')
net3_loss = np.genfromtxt('582 HW 4 data - net3 loss.csv', delimiter=',')
```

**Plots for different network sizes**


*   In-128-Out
*   In-64-64-Out
*   In-42-42-42-Out



```python
epochs = np.linspace(0,49,50)
plt.figure()
plt.plot(epochs, baseline_loss,'-o' , markersize = 3)
plt.plot(epochs, net2_loss, '-o' , markersize = 3)
plt.plot(epochs, net3_loss, '-o' , markersize = 3)
plt.title("Network Shapes: Loss vs Epoch", fontsize = 18)
plt.ylabel('Loss', fontsize = 16)
plt.xlabel('epoch', fontsize = 16)
plt.legend(['In-128-Out','In-64-64-Out','In-42-42-42-Out'], fontsize = 16)
plt.savefig('/content/drive/My Drive/582 HW 4/shapes_loss.png')
plt.show()
```

```python
plt.figure()
plt.errorbar(epochs, baseline_acc[:,0], baseline_acc[:,1], fmt = '-o', markersize = '3')
plt.errorbar(epochs, net2_acc[:,0], net2_acc[:,1], fmt = '-o', markersize = '3')
plt.errorbar(epochs, net3_acc[:,0], net3_acc[:,1], fmt = '-o', markersize = '3')
plt.title("Network Shapes: Validation Accuracy vs Epoch", fontsize = 18)
plt.ylabel('Validation Accuracy', fontsize = 16)
plt.xlabel('epoch', fontsize = 16)
plt.legend(['In-128-Out','In-64-64-Out','In-42-42-42-Out'], fontsize = 16)
plt.ylim([.6,.9])
plt.savefig('/content/drive/My Drive/582 HW 4/shapes_val.png')
plt.show()
```

Plots for different Optimizers (All other Hyperparameters the same)

```python
epochs = np.linspace(0,49,50)
plt.figure()
plt.plot(epochs, baseline_loss,'-o' , markersize = 3)
plt.plot(epochs, adadelta_loss,'-o' , markersize = 3)
plt.plot(epochs, adam_loss,'-o' , markersize = 3)
plt.title("Optimizers: Loss vs Epoch", fontsize = 18)
plt.ylabel('Loss', fontsize = 16)
plt.xlabel('epoch', fontsize = 16)
plt.legend(['SGD','AdaDelta','Adam'], fontsize = 16)
plt.savefig('/content/drive/My Drive/582 HW 4/optimizer_loss.png')
plt.show()
```

```python
plt.figure()
plt.errorbar(epochs, baseline_acc[:,0], baseline_acc[:,1], fmt = '-o', markersize = '3')
plt.errorbar(epochs, adadelta_acc[:,0], adadelta_acc[:,1], fmt = '-o', markersize = '3')
plt.errorbar(epochs, adam_acc[:,0], adam_acc[:,2], fmt = '-o', markersize = '3')
plt.title("Optimizers: Validation Accuracy vs Epoch", fontsize = 18)
plt.ylabel('Validation Accuracy', fontsize = 16)
plt.xlabel('epoch', fontsize = 16)
plt.legend(['SGD','AdaDelta','Adam'], fontsize = 16)
plt.ylim([.6,.92])

plt.savefig('/content/drive/My Drive/582 HW 4/optimizer_val.png')
plt.show()
```

Plotting the other stuff

```python
import numpy as np
import matplotlib.pyplot as plt

xavier_acc = np.genfromtxt('582 HW 4 data - Xavier_acc.csv', delimiter=',')
xavier_loss = np.genfromtxt('582 HW 4 data - Xavier_loss.csv', delimiter=',')
base_acc = np.genfromtxt('582 HW 4 data - baseline acc.csv', delimiter=',')
base_loss = np.genfromtxt('582 HW 4 data - baseline loss.csv', delimiter=',')
batch_acc = np.genfromtxt('582 HW 4 data - batch_acc.csv', delimiter=',')
batch_loss = np.genfromtxt('582 HW 4 data - batch_loss.csv', delimiter=',')
```

```python
epochs = np.linspace(0,49,50)
plt.figure()
plt.plot(epochs, base_loss,'-o' , markersize = 3)
plt.plot(epochs, xavier_loss,'-o' , markersize = 3)
plt.plot(epochs, batch_loss,'-o' , markersize = 3)
plt.title("Xavier and Batch Norm: Loss vs Epoch", fontsize = 18)
plt.ylabel('Loss', fontsize = 16)
plt.xlabel('epoch', fontsize = 16)
plt.legend(['Baseline','Xavier','Batch Norm'], fontsize = 16)
plt.savefig('/content/drive/My Drive/582 HW 4/extra_loss.png')
plt.show()
```

```python
plt.figure()
plt.errorbar(epochs, base_acc[:,0], base_acc[:,1], fmt = '-o', markersize = '3')
plt.errorbar(epochs, xavier_acc[:,0], xavier_acc[:,1], fmt = '-o', markersize = '3')
plt.errorbar(epochs, batch_acc[:,0], batch_acc[:,1], fmt = '-o', markersize = '3')
plt.title("Xavier and Batch Norm: Validation vs Epoch", fontsize = 18)
plt.ylabel('Validation Accuracy', fontsize = 16)
plt.xlabel('epoch', fontsize = 16)
plt.legend(['Baseline','Xavier','Batch Norm'], fontsize = 16)
plt.ylim([.6,.92])

plt.savefig('/content/drive/My Drive/582 HW 4/extra_val.png')
plt.show()
```
