**Random Walk Project**

Below, I implement a random walk function which takes a unit step in any direction from 0 to 2pi in the x-y plane.

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def randomWalks2D (numSteps):

  #set up a matrix of zeros to store the locations of the random walk
  locations = np.zeros((numSteps+1, 2))

  #take steps and store locations

  for i in range(1, numSteps+1):
    theta = random.random()*2* np.pi #random theta between 0 and 2 pi
    move = np.array([np.cos(theta), np.sin(theta)])
    locations[i] = locations[i-1] + move

  return locations
```

```python
walk = randomWalks2D(20)
plt.plot( walk[:,0], walk[:,1], alpha=0.5 )
plt.plot([0], [0],'r.')
plt.xlim(-10,10)
plt.ylim(-10,10);
```

1) a) What is the average distance of the random walk from the origin after n steps? Make a
plot showing how the average distance of the random walk from the origin depends on
n, and find a function that roughly fits your plot.


```python
def avg_dist(numWalks, numSteps):
  sum = 0 
  for i in range(numWalks):
    walk = randomWalks2D(numSteps)
    sum = sum + np.linalg.norm(walk[-1])

  return sum/numWalks
```

```python
list_of_avg_dists = [avg_dist(500, n) for n in range(1, 100)]
```

```python
plt.plot(list_of_avg_dists)
plt.xlabel("Number of Steps")
plt.ylabel("Average Ending Distance");
```

This looks like it could be a square root graph. To check, I will square the values to see if I get a straight line and find the line of best fit.

```python
sq_vals = [d**2 for d in list_of_avg_dists]
```

```python
plt.plot(sq_vals)
plt.xlabel("Number of Steps")
plt.ylabel("Average Ending Distance^2");
```

Looks pretty linear! I will find the line of best fit.

```python
from scipy import stats

result = stats.linregress(range(1,100), sq_vals)
result
```

```python
m = result.slope
b = result.intercept
```

```python
rootFit = [np.sqrt(m*x+b) if m*x+b > 0 else 0 for x in range(1,100)]
plt.plot(range(1,100), rootFit,'r-')
plt.plot(range(1,100), list_of_avg_dists, 'b-')
plt.xlabel("Number of Steps")
plt.ylabel("Average Ending Distance");
```

y = sqrt(0.781 n - 0.0633) fits the data very well, as we can see above.

(b) Since the walk is not on a grid, its very unlikely that it will return exactly to (0, 0).
However, it may return to within a circle of radius 1/2
around the origin. What proportion
of n-step random walks return at least once to within 1/2
unit of the origin? How does
this proportion depend on n? Investigate for various n and make a plot of your results.
Do you think that all random walks will eventually return near the origin? Why or why
not?

```python
def return_to_origin(walk):
  for i in walk[1:]:
    if np.linalg.norm(i) < 1/2:
      return True
  return False
```

```python
def prop_returns(numWalks, numSteps):
  sum = 0
  for _ in range(numWalks):
    walk = randomWalks2D(numSteps)
    if return_to_origin(walk) == True:
      sum = sum + 1
  return sum/ numWalks

```

prop_return seems to be working properly!

```python
list_of_prop_returns = [prop_returns(500, n) for n in range(1, 300)]
```

```python
plt.plot(list_of_prop_returns, 'b.');
plt.xlabel("Step Number")
plt.ylabel("Proportion Returning to Origin");
```

2) What happens if you constrain the random walk to the region 5  y  5? Modify your
code to keep the y-coordinate of the walk between 5 and 5. You may decide exactly how to

To keep the random walk within the region of -5 < y < 5, I will reroll the random steps if the step would make the simulation go outside of the bounds. A rerolled step will not count as a step, so if the number of steps is 10, there will be 10 steps within the bounds.

```python
def constrained_randomWalks2D (numSteps):

  #set up a matrix of zeros to store the locations of the random walk
  locations = np.zeros((numSteps+1, 2))

  #initialize counter
  i = 1


  #changed the for loop to a while loop
  while i <=numSteps:
    theta = random.random()*2* np.pi
    move = np.array([np.cos(theta), np.sin(theta)])
    yloc = locations[i - 1, 1] + np.sin(theta)
    if yloc > -5 and yloc < 5:
      locations[i] = locations[i-1] + move
      i = i + 1 #only increase the counter if a new location is updated



  return locations
```

3) For an exellent rating, I can investigate another aspect of the random walks. I will investigate the average diameter for both the constrained and unconstrained random walks. I expect this will be different than the average distance which I have looked at above, because diameter is the largest distance away from the center for a given walk. Maybe the constrained system will actually have a larger average diameter than the unconstrained system.

```python
 # get the largest distance away in a given walk
 def diameter(walk):
  dia_list = np.zeros(len(walk))
  for i in range(len(walk)):
     dia_list[i] = np.linalg.norm(walk[i])
  return max(dia_list)

# avg_diameter where I can choose which random walk I want, constrained or unconstrained
def avg_diameter(numSteps, numWalks, constraint):
  sum = 0
  for _ in range (numWalks):
    if constraint== True:
      walk = constrained_randomWalks2D(numSteps)
    if constraint== False:
      walk = randomWalks2D(numSteps)
      
    sum = sum + diameter(walk)

  return sum/numWalks
```
